```python
import pickle
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from Settings.SettingsKi67 import Settings as Settings
import Settings.GeneralSettings as generalSettings

settings = Settings(generalSettings) 
```

```python
atribute_types = pickle.load(open(settings.results_folder + "Fuzzification Statistics.p", "rb"))
```

```python
atributes = atribute_types.loc[(atribute_types.Gausses == 5)].sort_values(by=["Features type"])
# (atribute_types["Searched Class"] == "blue")
```

```python
atributes
```

```python
sns.set(style="whitegrid")
g = sns.catplot(x="Features type", y="Average Recall", hue="Searched Class", data=atributes, kind="bar", palette="muted", height=10, aspect=2)
g.despine(left=True)
g.set_ylabels("Average Recall")
```
