```python
import pickle
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from Settings.SettingsKi67 import Settings as Settings
import Settings.GeneralSettings as generalSettings

settings = Settings(generalSettings) 
```

#### Features Analysis

```python
atribute_types = pickle.load(open(settings.results_folder + "Features Analysis.p", "rb"))
atribute_types = atribute_types.sort_values(by=["Features type"])
display(atribute_types)

sns.set(style="whitegrid")
g = sns.catplot(x="Features type", y="Average Recall", hue="Searched Class", data=atribute_types, kind="bar", palette="muted", height=10, aspect=2)
g.despine(left=True)
g.set_ylabels("Average Recall")
```

#### Input Function Analysis

```python
input_functions_1 = pickle.load(open(settings.results_folder + "Input Functions Analysis 1.p", "rb"))
input_functions_1 = input_functions_1.sort_values(by=["Gausses"])

input_functions_2 = pickle.load(open(settings.results_folder + "Input Functions Analysis 2.p", "rb"))
input_functions_2 = input_functions_2.sort_values(by=["Gausses"])

input_functions = pd.concat([input_functions_1, input_functions_2])
input_functions = input_functions.reset_index(drop=True)
display(input_functions)

# sns.set(style="whitegrid")
# g = sns.catplot(x="Features type", y="Average Recall", hue="Searched Class", data=atribute_types, kind="bar", palette="muted", height=10, aspect=2)
# g.despine(left=True)
# g.set_ylabels("Average Recall")
```

```python
input_functions_gauss = input_functions.loc[input_functions.groupby(["Gausses", "Searched Class"])["Average Recall"].idxmax()] 
display(input_functions_gauss)
sns.set(style="whitegrid")
g = sns.catplot(x="Gausses", y="Average Recall", hue="Searched Class", data=input_functions_gauss, kind="bar", palette="Paired", height=8)
g.despine(left=True)
g.set_ylabels("Average Recall")
```

```python
input_functions_adj = input_functions.loc[input_functions.groupby(["Adjustment", "Searched Class"])["Average Recall"].idxmax()] 
display(input_functions_adj)
sns.set(style="whitegrid")
g = sns.catplot(x="Adjustment", y="Average Recall", hue="Searched Class", data=input_functions_adj, kind="bar", palette="Paired", height=4)
g.despine(left=True)
g.set_ylabels("Average Recall")
```

```python
input_functions_style = input_functions.loc[input_functions.groupby(["Style", "Searched Class"])["Average Recall"].idxmax()] 
display(input_functions_style)
sns.set(style="whitegrid")
g = sns.catplot(x="Style", y="Average Recall", hue="Searched Class", data=input_functions_style, kind="bar", palette="Paired", height=4)
g.despine(left=True)
g.set_ylabels("Average Recall")
```
