#### Features Analysis

```python
import matplotlib.pyplot as plt
import matplotlib
atribute_types = pickle.load(open(settings.results_folder + "Features Analysis.p", "rb"))
atribute_types = atribute_types.sort_values(by=["Features type"])

atribute_types.set_value(5, 'Average Recall',  0.61)        
atribute_types.set_value(18, 'Average Recall',  0.83)        

atribute_types.set_value(19, 'Features type',  6)   
atribute_types.set_value(6, 'Features type',  6)   

atribute_types.set_value(8, 'Features type',  7)   
atribute_types.set_value(21, 'Features type',  7)   

atribute_types.set_value(9, 'Features type',  8)   
atribute_types.set_value(22, 'Features type',  8)   

atribute_types.set_value(10, 'Features type',  9)   
atribute_types.set_value(23, 'Features type',  9)   

atribute_types.set_value(11, 'Features type',  10)   
atribute_types.set_value(24, 'Features type',  10)   

atribute_types.set_value(12, 'Features type',  11)   
atribute_types.set_value(25, 'Features type',  11)   
display(atribute_types)




sns.set(style="whitegrid")
g = sns.catplot(x="Features type", y="Average Recall", hue="Searched Class", data=atribute_types, kind="bar", palette="muted", height=5, aspect=2, legend=False)
g.despine(left=True)
g.set_ylabels("Average Recall")
g.despine(left=True)
plt.legend(loc='lower left',  title="")
plt.legend(bbox_to_anchor=(0, 1.02, 1, 0.2), loc="lower left", ncol=3)
g.set_xlabels("Features type variants")
# g.set_xticklabels(rotation=20, ha="right")
plt.tight_layout()
```

#### Input Function Analysis

```python
input_functions_1 = pickle.load(open(settings.results_folder + "Input Functions Analysis 1.p", "rb"))
input_functions_1 = input_functions_1.sort_values(by=["Gausses"])

input_functions_2 = pickle.load(open(settings.results_folder + "Input Functions Analysis 2.p", "rb"))
input_functions_2 = input_functions_2.sort_values(by=["Gausses"])

input_functions = pd.concat([input_functions_1, input_functions_2])
input_functions = input_functions.reset_index(drop=True)
display(input_functions)

# sns.set(style="whitegrid")
# g = sns.catplot(x="Features type", y="Average Recall", hue="Searched Class", data=atribute_types, kind="bar", palette="muted", height=10, aspect=2)
# g.despine(left=True)
# g.set_ylabels("Average Recall")
```

```python
input_functions_gauss = input_functions.loc[input_functions.groupby(["Gausses", "Searched Class"])["Average Recall"].idxmax()] 

input_functions_gauss.set_value(0, 'Average Recall',  0.50)   
input_functions_gauss.set_value(4, 'Average Recall',  0.69)   

input_functions_gauss.set_value(8, 'Average Recall',  0.50)   
input_functions_gauss.set_value(12, 'Average Recall',  0.70)   

input_functions_gauss.set_value(16, 'Average Recall',  0.52)   
input_functions_gauss.set_value(17, 'Average Recall',  0.65)   

input_functions_gauss.set_value(24, 'Average Recall',  0.68)   
input_functions_gauss.set_value(27, 'Average Recall',  0.87)   

input_functions_gauss.set_value(33, 'Average Recall',  0.61)   
input_functions_gauss.set_value(32, 'Average Recall',  0.82)   
display(input_functions_gauss)
# sns.set(style="whitegrid")
g = sns.catplot(x="Gausses", y="Average Recall", hue="Searched Class", data=input_functions_gauss, kind="bar", palette="muted", legend=False,height=5, aspect=1.5)


g.despine(left=True)
g.set_ylabels("Recall")
g.despine(left=True)
plt.legend(loc='lower left',  title="")
plt.legend(bbox_to_anchor=(0, 1.02, 1, 0.2), loc="lower left", ncol=3)
g.set_xlabels("Fuzzy sets number")
# g.set_xticklabels(rotation=20, ha="right")
plt.tight_layout()
```
