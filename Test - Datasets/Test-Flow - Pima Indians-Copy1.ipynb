```python
import sympy as sy
from sympy import symbols, Eq, solve

```

```python
sigma = symbols('sigma', real=True)
eq1 = Eq(sy.exp(-((0.25-0.4)**2.)/sigma**2.) - 0.5)

res = solve((eq1), (sigma))
print(res)
adam = []
for x in res:
    if x >= 0:
        adam.append(x)

print(adam)
# return np.float64(x_value), np.float64(sigma_value)
```
